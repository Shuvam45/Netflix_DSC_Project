```python
df['Release_Date']=pd.to_datetime(df['Release_Date'])
df['Release_Date'].dtype
```

```python
df['Release_Date']=df['Release_Date'].dt.year
df['Release_Date'].dtype
```

```python
df.head()
```

```python
### categorize avg votes column
def categorize(df, col, labels):
  edges= [df[col].describe()['min'],
          df[col].describe()['25%'],
          df[col].describe()['50%'],
          df[col].describe()['75%'],
          df[col].describe()['max']]

  df[col]=pd.cut(df[col], edges, labels=labels, duplicates='drop')
  return df
```

```python
labels=['not popular', 'below avg', 'average', 'popular']
categorize(df, 'Vote_Average', labels)
df['Vote_Average'].unique()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

Split genres into list and explode dataframe to ave only one genre per row for each movie


```python
df['Genre'] = df['Genre'].astype(str).str.split(', ')
df=df.explode('Genre').reset_index(drop=True)
df.head()
```
